```python
import bs4 as BeautifulSoup
import pandas as pd
```

```python
f = ''.join(open('/Users/franciscojavierarceo/Downloads/nj_schools.html', 'r').readlines())
doc = BeautifulSoup.BeautifulSoup(f)
```

```python
ranks = doc.select('ol[class="item-list__OrderedListStyled-sc-18yjqdy-0 kFQUBh"]')[0]
schools = ranks.select('li[class="item-list__ListItemStyled-sc-18yjqdy-1 boZDDO"]')
school = schools[0]
```

```python
school
```

```python
t1 = 'a[class="DetailCardHighSchools__AnchorStyled-zfywu8-7 gXAAll"]'
t2 = 'span[class="with-icon__Rank-sc-1spb2w-2 hEfLoa"]'
t3 = 'p[class="Paragraph-sc-1iyax29-0 iTUzJy"]'
t4 = 'span[class="RankingList__RankStyled-sc-7e61t7-1 WwFwZ"]'
tags = {'t1': t1, 't2': t2, 't3': t3, 't4': t4}

def get_record(school, config=tags):
    record = {
        'Town': school.select(config['t3'])[0].text,
        'School Name': school.select(config['t1'])[0].text,
        'NJ Ranking': school.select(config['t2'])[0].text.strip().replace('#',''),
        'US Ranking': school.select(config['t4'])[1].text.strip().replace('#',''),
    }
    return record
```

```python
output = []
for i, school in enumerate(schools):
    try:
        output.append(get_record(school))
    except:
        print(f'record {i} failed')
```

```python
sdf = pd.DataFrame(output)
```

```python
search_engine = uszipcode.SearchEngine()
```

```python
res = search_engine.by_city_and_state(city='Edison', state='New Jersey')[0]
```

```python
zoutput = []
for i, row in sdf.iterrows():
    city = row['Town'].replace(", NJ", "")
    try:
        res = search_engine.by_city_and_state(city=city, state='New Jersey')[0]
        zoutput.append(res.to_dict())
    except:
        print(f'no result for {city}')
```

```python
zdf = pd.DataFrame(zoutput)
zdf.drop('common_city_list', axis=1, inplace=True)
zdf.drop_duplicates(inplace=True)
```

```python
zdf.head()
```

```python
finaldf = sdf.merge(zdf, how='left', left_on='Town', right_on='post_office_city')
```

```python
res2 = res.to_dict()
res2.pop('common_city_list')

for k in res2:
    finaldf.loc[finaldf["School Name"] == "West Windsor-Plainsboro High School South", k] = res2[k]
```

```python
def manually_adjust(df, school, zipcode):
    res = search_engine.by_zipcode(zipcode).to_dict()
    res.pop('common_city_list')

    for k in res:
        df.loc[df["School Name"] == school, k] = res[k]
```
